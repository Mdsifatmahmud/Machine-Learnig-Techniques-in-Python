```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv(r'canada_per_capita_income.csv')
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.title('Per Capita Income in Canada')
plt.scatter(df['year'], df['per capita income (US$)'], color='green', marker='o')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']], df['per capita income (US$)'])
```

```python
reg.predict([[2020]])
reg.predict([[2025]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
# Create linear regression object
# Fit the model
# y = mx + b
# y = 45431.01947053 * x + (-163219.2531122)
45431.01947053*2020 + (-163219.2531122)
```

Save Model Using Joblib And Pickle

```python
# Save the model to disk
import pickle
```

```python
with open('model_pickle', 'wb') as f:
    pickle.dump(reg, f)
```
